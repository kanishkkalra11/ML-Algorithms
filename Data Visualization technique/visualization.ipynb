Create dataset for visualization

Visualisation: Displaying Female Labour Participation vs GDP per Capita for 6 countries(India, Pakistan, China, Russia, United States, Australia) from 1990 to 2010. The countries are grouped according to their economic and political ideologies during the cold war. India and Pakistan are third world countries (shown in gray), US and Australia are first world countries (shown in green), and China and Russia are second world countries (shown in blue). The size of the datapoints are indiactive of the population of the corresponding countries.

The above data are taken from https://www.gapminder.org/data/.

```
countries = ['India','Pakistan','Australia','China','Russia','United States']
worlds = ['third','third','first','second','second','first']
```

```
female_labour = []
for country in countries:
    female_labour.append(a.loc[a['country']==country])
```

```
gdp = []
for country in countries:
    gdp.append(b.loc[b['country']==country])
```

```
pop = []
for country in countries:
    pop.append(c.loc[c['country']==country])
```

```
Frame = {}
initial = 1990
while(initial<=2010):
    Frame[initial] = {}
    Frame[initial]['country'] = countries
    Frame[initial]['population'] = [int(pop[i][str(initial)]) for i in range(len(pop))]
    Frame[initial]['gdpcapita'] = [int(gdp[i][str(initial)]) for i in range(len(gdp))]
    Frame[initial]['femalelabourparticipation'] = [int(female_labour[i][str(initial)]) for i in range(len(female_labour))]
    Frame[initial]['worlds'] = worlds
    initial+=1
```

```
for key in Frame.keys():
    Frame[key] = pd.DataFrame(Frame[key])
```

Plotting

```
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
sns.set_style("white")
import pandas as pd
my_dpi=96
```

```
for year in range(1990,2011):
    Frame[year]['worlds']=pd.Categorical(Frame[year]['worlds'])
    fig = plt.figure(figsize=(680/my_dpi, 480/my_dpi), dpi=my_dpi)
    tmp=Frame[year]
    plt.scatter(tmp['femalelabourparticipation'], tmp['gdpcapita'] , s=tmp['population']/200000 , c=tmp['worlds'].cat.codes, cmap="Accent", alpha=0.6, edgecolors="white", linewidth=2)
    plt.yscale('log')
    plt.xlabel("Female Labour Participation (%)")
    plt.ylabel("GDP per Capita")
    plt.title("Year: "+str(year) )
    plt.ylim(0,100000)
    plt.xlim(10, 90)

    filename='Visual'+str(year)+'.png'
    plt.savefig(filename, dpi=96)
    plt.gca()
```
